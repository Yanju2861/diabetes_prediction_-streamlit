```python
# For visualiztion
import  matplotlib.pyplot as plt
# For Numpy arrays
import numpy as np
import pandas as pd

# Import seaborn
import seaborn as sns ## import seaborn as sb
%matplotlib inline
```

```python
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
print(type(tips))
print(tips.shape)
print(tips.columns)
tips
```

### 1.2 First, analyse the dataset 

```python
### Do this for any dataset after loading it
print(tips.head(2))
print(tips.tail(2))
tips.info()

### OR Check datatypes using dtypes
# tips.dtypes
```

```python
### Check for data value counts
print("Value_Counts for sex column \n", tips['sex'].value_counts())

### Check for unique values
X = tips['time'].unique()
print(type(X))
print(X)
for each in X:
    print(each)
```

# 2. Plots made easy with Seaborn

### 2.1 Scatter Plot for checking relationship among two (or more) variables

```python
## Scatter Plot 
## Bivariate Distribution
sns.relplot (x = tips["total_bill"], y = tips["tip"])  #data=tips, NOT GOOD

sns.relplot (data= tips, x="total_bill", y="tip")  #data=tips, GOOd
```

```python
# Create a visualization
## Bivariate Distribution
sns.relplot(data=tips, x="total_bill", y="tip", height=6, aspect=1)
```

```python
# Create a visualization
sns.relplot (data=tips, x="total_bill", y="tip", col="smoker", hue='sex', style='sex', size='size' ) ###, hue='smoker' , style='day' ) #, size='size')
```

```python
# Create a visualization
sns.relplot (
    data=tips,
    x="total_bill", y="tip", size='size',height=5, aspect= 1, style='size', color = 'red')
```

```python
sns.relplot (
    data=tips,
    x="total_bill", y="tip", col="time",
     hue="sex", style="sex", size="size"
)   #### style="smoker",     hue="smoker", 
```

```python
sns.relplot (
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size"
)
```

### 2.2 HISTOGRAM for Univariate analysis of continuous variable

```python
# FOR HISTOGRAM
# Distribution Plot: Single variable, univariate distribution
sns.displot(data= tips, x= 'tip',bins=12,kde=True)

# We also get kde, a kernel density function (kde)
tips['tip'].skew()
```

```python
# We also get kde, a kernel density ESTIMATION (KDE) function
sns.displot(data= tips, x='size', bins = 6, kde=True)
```

```python
# FOR HISTOGRAM (different values for bin)
sns.displot(tips['total_bill'],bins = 15)
# We also get kde, a kernel density function
sns.displot(tips, x='total_bill', kde= True, bins = 30)
```

### 2.3 RUGPLOT for Univariate analysis of continuous variable

```python
# Rugplot 
# Similar to Tally marks
# sns.rugplot (tips['total_bill'])

sns.rugplot (data= tips, x='total_bill')
```

```python
# using height parameter
sns.rugplot (data= tips, x='total_bill', height=0.1 )
```

```python
### setting x and y parameter
sns.rugplot (data= tips, x='total_bill', y='tip', height=0.2 )
```

### 2.4 RUGPLOT with SCATTER PLOT for Bivariate analysis of continuous variables

```python
# Rugplot WITH SCATTER PLOT
# sns.rugplot (data=tips, y= 'tip', x= 'total_bill')
sns.rugplot (data=tips, y= 'tip', x= 'total_bill',height=0.1)
sns.scatterplot (data=tips, y= 'tip', x= 'total_bill')
```

```python
# Rugplot WITH SCATTER PLOT (with HUE parameter)
fig1, (ax1,ax2) = plt.subplots(1,2, figsize=(10, 6));

sns.rugplot (data=tips, ax= ax1, y= 'tip', x= 'total_bill', hue='time',height=0.1)
# sns.rugplot (data=tips, y= 'tip', x= 'total_bill',height=0.1)
sns.scatterplot (data=tips, ax= ax1, y= 'tip', x= 'total_bill', hue='time')

sns.rugplot (data=tips, ax= ax2, y= 'tip', x= 'size', hue='time',height=0.1)
# sns.rugplot (data=tips, y= 'tip', x= 'total_bill',height=0.1)
sns.scatterplot (data=tips, ax= ax2, y= 'tip', x= 'size', hue='time')
```

```python
# Rugplot WITH SCATTER PLOT  (with HUE parameter correct way)
fig1, ax1 = plt.subplots() ##    figsize=(10, 7));
sns.rugplot (data=tips, ax= ax1, y= 'tip', x= 'total_bill', hue='time',height=0.15)
# sns.rugplot (data=tips, y= 'tip', x= 'total_bill',height=0.1)
sns.scatterplot (data=tips, ax= ax1, y= 'tip', x= 'total_bill', hue='time')
```

### 2.5 KDEPLOT for Univariate analysis of continuous variable

```python
##  kernel density estimate (KDE) plot 
'''KDE represents the data using a continuous probability density curve in one or more dimensions, very similar to a histogram'''
sns.kdeplot(data=tips, x = 'total_bill', y='tip')
```

### 2.5 BARPLOT for analysis of numerical variables (estimation)

```python
# Create plots with categorical variables
### Univariate Distribution
sns.barplot ( data=tips, x= 'day', y='total_bill', estimator=np.sum)   # Mean is default estimator
```

```python
### Using two variables
# sns.barplot ( data=tips, x='sex', y= 'tip' )   # Mean is default estimator
sns.barplot (data=tips, x='sex', y= 'tip', estimator=np.sum )
```

```python
# sns.barplot (data=tips, x='sex', y= 'tip', hue='day' )   # Using hue parameter

# sns.barplot (data=tips, x='sex', y= 'tip', hue='day', estimator=np.min )  # Using other estimator
sns.barplot (data=tips, x='sex', y= 'tip', hue='day', estimator=np.max )  # Using other estimator
```

# 3. Bivariate Distribution Plots

### 3.1 JOINT PLOT for bivariate distibution 

```python
# Joint Plot: three plots, therein
sns.jointplot(data = tips, x= 'total_bill' , y= 'tip' ) # kind='scatter' default
```

```python
# Joint Plot: Other values of @kind are 'hex', 'reg', 'kde'
sns.jointplot(data = tips, x= 'total_bill' , y= 'tip' , kind = 'hex') # kind='scatter' default
```

```python
# Joint Plot: Adding @hue parameter 
sns.jointplot(data = tips, x= 'total_bill' , y= 'tip', hue='sex' ) 
```

### 3.1 PAIR PLOT for bivariate distribution among every possible pair of numeric variables 

```python
## PAIRPLOT
# For every pair of numerical values in dataset
# sns.pairplot(tips) 
sns.pairplot(tips[ ['total_bill','tip','size'] ])
```

```python
## PAIRPLOT: For selected pair of numerical values in dataset
sns.pairplot(tips, vars= ['tip','total_bill'], kind= 'reg', hue='sex')       ## kind = any of the ‘scatter’, ‘kde’, ‘hist’, ‘reg’
```

```python
# For every pair of numerical values in dataset
# Additional discrimination based on addtinal column values
sns.pairplot(tips, hue='day', palette = 'autumn') 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Create a heatmap
plt.subplot(1, 2, 1)
sns.heatmap(data, annot=True, cmap="YlGnBu", linewidths=0.5, square=True, cbar=True, fmt='d')
plt.title('Heatmap')

# Create a clustermap
plt.subplot(1, 2, 2)
sns.clustermap(data, annot=True, cmap="YlGnBu", linewidths=0.5, square=True, cbar=True, fmt='d')
plt.title('Clustermap')

plt.tight_layout()
plt.show()
```
