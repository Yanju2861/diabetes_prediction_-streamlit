# Web Scraping flipkart.com

```python
from bs4 import BeautifulSoup # Importing the BeautifulSoup package
import requests # Importing requests library
import csv
import pandas as pd 

### Title of any laptop- div._4rR01T
### Selling Price-  div._30jeq3._1_WHN1   
### Processor name- li.rgWa7D
### Discount- span
### Original Price- div._3I9_wc._27UcVY
### Ratings- div._3LWZlK      
```

```python
my_link = 'https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=fb6e74c4-a005-4471-96e8-c33c18c1308d&as-searchtext=laptop'
```

```python
req = requests.get(my_link) # Requesting the content of the URL
content = req.content # Getting the content
type(content)
```

```python
soup = BeautifulSoup(req.content,'html.parser')
# Here we need to specify the content variable and the parser which is the HTML parser
# So now soup is a variable of the BeautifulSoup object of our parsed HTML
```

```python
print(soup.prettify()) # Displays the whole code of the webpage
```

```python
tags = soup.find('div')
len(tags)
# tags
```

```python
desc = soup.find_all('div') 
len(desc)
# desc
```

```python
desc = soup.find_all('div' , class_='_4rR01T')    ##_4rR01T   _4rR01T
print(type(desc))
print(len(desc))
desc
```

```python
desc = soup.find_all('div' , class_='_4rR01T') 
# Extracting the descriptions from the website using the find method - grabbing the div tag which has the classname _3wU53n
# So now this returns all the div tags with the classname of _3wU53n
# As class is a special keyword in python we have to use the class_ keyword and pass the arguments here
descriptions = []
for i in range(len(desc)):
    descriptions.append(desc[i].text) # We can even access the child tags with dot access.
descriptions
# len(descriptions)
```

```python
commonclass = soup.find_all('li',class_='rgWa7D')   ### rgWa7D  ### rgWa7D
len(commonclass)
```

```python
commonclass = soup.find_all('li',class_='rgWa7D') 
# We observe that the classnames for the different specifications are under one div.
#So we need to apply some method to extract the different features.
 
### Create empty lists for the features
processors=[]
ram=[]
os=[]
storage=[]
inches=[]
warranty=[]

# sing_list = []

for i in range(0,len(commonclass)):
    p = commonclass[i].text # Extracting the text from the tags
    
  
    if("Core" in p or "Processor" in p): 
        processors.append(p)
    elif("RAM" in p ): # If RAM is present in the text then append it to the ram list. Similarly do this for the other features as well
        ram.append(p)
    elif("HDD" in p or 'SSD' in p):
        storage.append(p)
    elif("Operating" in p):
        os.append(p)
    elif("Display" in p or "inches" in p):
        inches.append(p)
    elif("Warranty" in p or "warranty" in p or "YEAR" in p or "Year" in p or "year" in p):
        warranty.append(p)
# len(sing_list)
# sing_list
```

```python
price = soup.find_all('div',class_='_30jeq3 _1_WHN1') # Extracting price of each laptop from the website by inspecting the element_30jeq3 _1_WHN1
prices = []                           ###  _30jeq3 _1_WHN1  _30jeq3 _1_WHN1
for i in range(len(price)):
    prices.append(price[i].text)  
prices
len(prices)

```

```python
prices
```

```python
original_price = soup.find_all('div',class_='_3I9_wc _27UcVY')   # For original price, it is "<div class="_3I9_wc _27UcVY">₹73,544</div>"
len(original_price)          ### _3I9_wc _27UcVY

```

```python
original_prices = []
for i in range(len(original_price)):
    original_prices.append(original_price[i].text)
original_prices
# len(original_prices) # Here we get 22 since there is no original price listed for 2 laptops-this means that there is actually no discount on the laptop and the original price is same as the discounted price
```

```python
rating = soup.find_all('div',class_='_3LWZlK') # Extracting the ratings of all the laptops _3LWZlK  _3LWZlK
ratings = []
for i in range(len(rating)):
    ratings.append(rating[i].text)
ratings
len(ratings) # We observe that the classname for the recommended laptops is also same as the present laptops so that's why it is extracting the ratings of recommned laptops as well
# This is increasing the number of ratings
```

```python
ratings
```

```python
discount = soup.find_all('span')
len(discount) # There are many span tags, but we need to find only the tags which contain discount 
discount
```

```python
discounts = []
my_next = []
for i in range(0,len(discount)):
    p=discount[i].text
    if("% off" in p): # Extracting the tags which contain the discount(all the discounts contain % off so checking with that condition we obtain all the laptops which have discounts)
        discounts.append(p)
#     elif ("NEXT" in p):
#         my_next.append(p)
        
    
discounts
len(discounts) 
```
